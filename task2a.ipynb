```python
# Install if needed:
# pip install transformers nltk

import nltk
from nltk.corpus import stopwords
from transformers import AutoTokenizer, BertTokenizer

# Download stopwords (only first time)
nltk.download('stopwords')

# -----------------------------
# Load BERT Tokenizer
# -----------------------------
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

text = "Hello, how are you doing today?"

print("Original Text:", text)
print("-" * 50)

# -----------------------------
# 1️⃣ Basic Tokenization
# -----------------------------
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)

token_ids = tokenizer.encode(text)
print("Token IDs:", token_ids)

encoded = tokenizer(text, max_length=64, truncation=True)
print("Full Encoding:", encoded)

print("-" * 50)

# -----------------------------
# 2️⃣ Remove NLTK Stopwords
# -----------------------------
stop_words = set(stopwords.words("english"))

filtered_words = [word for word in text.split() if word.lower() not in stop_words]
filtered_text = " ".join(filtered_words)

filtered_tokens = tokenizer.tokenize(filtered_text)
print("After Removing NLTK Stopwords:", filtered_tokens)

print("-" * 50)

# -----------------------------
# 3️⃣ Replace Stopwords with [STOP]
# -----------------------------
filtered_text_with_stop = ' '.join(
    [word if word.lower() not in stop_words else '[STOP]'
     for word in text.split()]
)

bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
tokens_with_stop = bert_tokenizer.tokenize(filtered_text_with_stop)

print("Stopwords Replaced with [STOP]:", tokens_with_stop)

print("-" * 50)

# -----------------------------
# 4️⃣ Custom Stopwords
# -----------------------------
custom_stopwords = ["hello", "are", "you", "doing", "today"]

custom_tokens = []
for word in text.split():
    if word.lower() not in custom_stopwords:
        custom_tokens.extend(tokenizer.tokenize(word))

print("After Removing Custom Stopwords:", custom_tokens)
```
